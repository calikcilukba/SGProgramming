```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation
import random

n_particle = 10
n_iter = 100
pos = []
lis = []
for i in range(n_particle):
    lis = []
    x = random.randint(1,20)
    y = random.randint(1,20)
    lis.append(x)
    lis.append(y)
    pos.append(lis)

x1 = [pos[0]]
x2 = [pos[1]]
x3 = [pos[2]]
x4 = [pos[3]]
x5 = [pos[4]]
x6 = [pos[5]]
x7 = [pos[6]]
x8 = [pos[7]]
x9 = [pos[8]]
x10 = [pos[9]]
# x1
```

```python
def perulangan(yt):
    j = 0
    gg = yt
    for i in range(n_iter):
        xa = []
        ya = []
        lisa = []
        r = np.random.uniform(0,1)
        if r > 0 and r <= 0.25:
            xa = gg[j][0] + 1
            ya = gg[j][1]
        elif r > 0.25 and r <= 0.5:
            xa = gg[j][0] - 1
            ya = gg[j][1]
        elif r > 0.5 and r <= 0.75:
            xa = gg[j][0]
            ya = gg[j][1] - 1
        elif r > 0.75 and r <= 1: 
            xa = gg[j][0]
            ya = gg[j][1] + 1
        j = j + 1
        lisa.append(xa)
        lisa.append(ya)
        gg.append(lisa)
        
    return yt

a1 = perulangan(x1)
a2 = perulangan(x2)
a3 = perulangan(x3)
a4 = perulangan(x4)
a5 = perulangan(x5)
a6 = perulangan(x6)
a7 = perulangan(x7)
a8 = perulangan(x8)
a9 = perulangan(x9)
a10 = perulangan(x10)

for i in range(n_iter):
    plt.scatter(a1[i][0], a1[i][1], c="pink")
    plt.scatter(a2[i][0], a2[i][1], c="yellow")
    plt.scatter(a3[i][0], a3[i][1], c="black")
    plt.scatter(a4[i][0], a4[i][1], c="blue")
    plt.scatter(a5[i][0], a5[i][1], c="cyan")
    plt.scatter(a6[i][0], a6[i][1], c="orange")
    plt.scatter(a7[i][0], a7[i][1], c="green")
    plt.scatter(a8[i][0], a8[i][1], c="red")
    plt.scatter(a9[i][0], a9[i][1], c="y")
    plt.scatter(a10[i][0], a10[i][1], c="m")
plt.show()
```
